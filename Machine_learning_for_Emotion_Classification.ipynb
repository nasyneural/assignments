### History of machine learning

Current advances in technology and machinery algorithms give many insights and abilities to diagnostics and research. One of such methods is machine learning, which is applicable to any type of data.

The term machine learning was proposed by Arthur Samuel in 1959. But models, implementing machine learning, have existed earlier - the first known is presented in Donald Hebb’s book ‘The Organisation of Behavior’ (1949), based on a model of brain cell interaction. Starting from the Hebbian model, machine learning took a notable place in modeling of cognitive processes. Hebb's model can be described as a way of altering the relationships between artificial neurons (also referred to as nodes) and the changes to individual neurons. The relationship between two neurons/nodes strengthens if the two neurons/nodes are activated at the same time and weakens if they are activated separately.

Later came the model that embodied Machine Learning the Game of Checkers, developed by Arthur Samuel. In order to overcome the problem of the extremely small amount of memory in a computer, Samuel initiated what is called alpha-beta pruning. His design included a scoring function using the positions of the pieces on the board. The scoring function attempted to measure the chances of each side winning. The next move was chosen using a minimax strategy, which evolved into the minimax algorithm.

Another early example of machine learning was the development of the perceptron, which is a type of artificial neural network used for binary classification. The model was created in the 1950s by Frank Rosenblatt, who combined Donald Hebb’s model of brain cell interaction with Arthur Samuel’s machine learning efforts. Futher this model has been widely used in various applications, such as image and speech recognition (Ahad et al., 2002; Lai & Deng, 2018).

In the 1980s and 1990s, there was a surge in the development of new machine learning algorithms and techniques, including support vector machines (SVMs), random forest and artificial neural networks. These algorithms have been widely used in various applications, such as natural language processing (Rameshbhai, 2019), object recognition (Schrimpf et al., 2020), and machine translation (De Baets & De Sutter, 2022).

Overall, the history of machine learning classification has been marked by the development of increasingly sophisticated algorithms and techniques, which have been used to solve a wide range of scientific and applied problems. 

### Main implications of machine learning for science

*Predictive modeling:* Machine learning can be used to build models that can predict the outcomes of certain scientific experiments or processes, based on past data. For example, machine learning can be used to predict the reinforceement learning behavior (Snutton & Barto, 1998), risk of diseases (Uddin et al., 2019; Bind et al., 2015; Caballero et al., 2021) can help with investigation of the neural basis of decision-making and other complex behaviors (Peterson et al., 2021).

*Also, it can be used to predict chemical reactions (Meuwly, 2021; Hein, 2021), properties and efficiency of new drugs (Gerdes et al., 2021; Patel et al., 2020), and even climate change (Rolnick, 2022).*

*Data analysis:* Machine learning provides novel opportunities for studying complex relationships among phenomena and representing causality among multiple variables. It can help to identify patterns and trends in large scales of data, demanding enormous time of manual work, as well as make predictions about future outcomes. For example, it can give useful insights for diagnostics:  distinguish epileptic seizures (Siddiqui, 2020; Usman et al., 2018) and classify other diseases (Bratić, 2018; Zhang, 2019).

It supports the role of machine learning in cognitive science as a tool for the validation of psychological models and theoretical constructs that is highly versatile and could lead to many new insights (Krumpe et al., 2018). Various cognitive states, as well as diseases, evidentially alter electrical activity of the brain (Mueller et al. (2010)), and this gives a remarkable opportunity for application of machine learning algorithms for the classification of ERP data.

**Logistic Regression**

Logistic regression is used to predict the probability of a data point belonging to a particular class. It is a linear model that works by fitting a linear boundary to the data, such that the data points on one side of the boundary belong to one class, and the data points on the other side belong to the other class.

**Naive Bayes**

Naive Bayes is a probabilistic algorithm that is used for classification tasks. It is based on Bayes' theorem, which is a mathematical principle that describes the relationship between probability and conditional probability.

The basic idea behind the Naive Bayes classification algorithm is to predict the class of a data point based on the likelihood of the data point belonging to each class, given the features or characteristics of the data point. This can be written as follows:

**K-Nearest Neighbors (KNN)**

The K-Nearest Neighbors is a non-parametric machine learning algorithm. It works by finding the K nearest data points to a given data point, and then assigning the given data point to the class that is most common among the K nearest neighbors.

The distance between data points is typically calculated using a distance measure, such as Euclidean distance or cosine similarity. The number of nearest neighbors (K) is a hyperparameter that is chosen by the user.

tg_image_1978774684.jpeg

To classify a new data point, the KNN classification algorithm calculates the distances between the new data point and all the data points in the training set, and then selects the K data points with the smallest distances. The new data point is then assigned to the class that is most common among the K nearest neighbors.

**Decision Tree**

The Decision Tree is a non-parametric algorithm that works by building a tree-like structure, where each node in the 'tree' represents a feature or characteristic of the data, and the branches represent the possible values of that feature. The leaves of the tree represent the class labels.

To classify a new data point, the Decision Tree classification algorithm starts at the root node of the tree and traverses the tree by following the branches that correspond to the values of the features of the data point. The algorithm stops at a leaf node, and the class label associated with that leaf node is the predicted class for the data point.

The Decision Tree classification algorithm uses an optimization criterion, such as entropy or the Gini index, to determine the best split at each node. The optimization criterion measures the purity of the data at each node, and the goal is to find splits that result in the purest subsets of data.

**Random Forest**

Random Forest works by building a large number of decision trees, and then combining the predictions of the individual trees to make a final prediction.

Each decision tree in a Random Forest is built using a random subset of the data, and a random subset of the features of the data. This process is repeated multiple times, resulting in a large number of decision trees that are all trained on different subsets of the data.

To classify a new data point, the Random Forest classification algorithm makes a prediction using each of the individual decision trees, and then combines the predictions using a voting process. The class label that receives the most votes is the final predicted class for the data point.

**Support Vector Machines**

The SVM  algorithm tries to find the hyperplane that has the maximum margin, which is the distance between the hyperplane and the closest data points from each class. The data points that are closest to the hyperplane are called support vectors, and they define the position of the hyperplane.

#### Main steps to implement machine learning classification: 

1. *Preprocessing* involves cleaning and preparing the EEG data for analysis. This may include steps such as noise removing, artifact correction, and filtering the data.

2. *Feature extraction* involves extracting relevant features or characteristics from the EEG data that can be used as input to the machine learning classifier. This may involve using techniques such as time-frequency analysis, spatial filtering, principal component analysis (PCA), independent component analysis (ICA) and auto-regression models (AR) (Delorme et al., 2007; Garrett et al., 2003; Peters et al., 2001).

3. *Model training* involves using the extracted features to train a machine learning classifier. This may involve splitting the data into training and test sets, selecting an appropriate model and hyperparameters, and training the model using the training data.

4. *Model evaluation* involves evaluating the performance of the trained model on the test data. This may involve calculating metrics such as accuracy, precision, and recall, and comparing the performance of different models.

```python
column_list = emo_dataframe.columns.tolist()
print(column_list[:-1])
```

```python
class_dfs = []
for name in names:
    class_dfs.append(emo_dataframe[emo_dataframe['label'] == name][column_list[:-1]])
```

```python
f_vals, p_vals = stats.f_oneway(*class_dfs)

for i in range(len(column_list[:-1])):
    print(f"Column {column_list[:-1][i]}: F-value = {f_vals[i]:.3f}, p-value = {p_vals[i]:.3f}")
```

```python
significant_columns = [column_list[:-1][i] for i in range(len(column_list[:-1])) if p_vals[i] < 0.0001]

print(significant_columns)
```

```python
len(significant_columns)
```

```python
def Transform_data(data):
    #Encoding Lables into numbers
    encoding_data = ({'NEUTRAL': 0, 'POSITIVE': 1, 'NEGATIVE': -1} )
    data_encoded = data.replace(encoding_data)
    #getting brain signals into x variable
    x=data_encoded[significant_columns]
    #getting labels into y variable
    y = data_encoded.loc[:,'label'].values
    scaler = StandardScaler()
    #scaling Brain Signals
    scaler.fit(x)
    X = scaler.transform(x)
    return X,y
```

```python
#Calling above function and splitting dataset into train and test
X,y = Transform_data(emo_dataframe)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 22, shuffle=True)
```

```python
classifiers = [
    RandomForestClassifier(),
    LogisticRegression(),
    SVC(),
    KNeighborsClassifier(),
    GaussianNB(),
    DecisionTreeClassifier()
]
```

```python
# Train each classifier and store the results

accuracy = {}
info = {}

for clf in classifiers:
    name = clf.__class__.__name__
    clf.fit(X_train, y_train)

    train_acc = clf.score(X_train, y_train)
    test_acc = clf.score(X_test, y_test)

    accuracy[f'{name}_train'] = round(train_acc*100,3)
    accuracy[f'{name}_test'] = round(test_acc*100,3)

    
    # Make predictions on the test set
    y_pred = clf.predict(X_test)

    # Calculate the confusion matrix
    cm = confusion_matrix(y_test, y_pred)

    sensitivity = cm[1, 1] / np.sum(cm[1, :])
    specificity = cm[0, 0] / np.sum(cm[0, :])
    precision = cm[1, 1] / np.sum(cm[:, 1])
    recall = sensitivity

    info[name] = {
        'sensitivity': sensitivity,
        'specificity': specificity,
        'precision': precision,
        'recall': recall
    }


    # Plot the confusion matrix
    plt.imshow(cm, cmap='viridis')
    plt.colorbar()
    plt.xlabel('Predicted label')
    plt.ylabel('True label')
    plt.title(f'Confusion Matrix for {name}')
    plt.show()
    
```
